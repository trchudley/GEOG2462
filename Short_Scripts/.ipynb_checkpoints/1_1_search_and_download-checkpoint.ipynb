# AOI

```python
lat, lon = 54.774504, -1.575872
size = 10000
```

```python
point = ee.Geometry.Point(lon, lat)
region = point.buffer(size/2).bounds()
```

```python
Map = geemap.Map()#center=(x, y), zoom=12)
Map.addLayer(region, {}, "Search region")
Map.centerObject(region)
Map
```

# Search and clip

```python
date_begin = '2023-05-01'
date_end = '2023-09-30'

# Get Landsat 8 image collection
landsat8_collection = ee.ImageCollection("LANDSAT/LC08/C02/T1_TOA")

# Filter to desired region and date bounds
landsat8_collection = landsat8_collection.filterBounds(region)
landsat8_collection = landsat8_collection.filterDate(date_begin, date_end)

# Select bands: red, green, blue, NIR, SWIR
landsat8_collection = landsat8_collection.select(['B4', 'B3', 'B2', 'B5', 'B6'])
```

```python
# landsat8_collection.select('B8')
```

```python
image = landsat8_collection.sort('CLOUD_COVER').first()
image = image.clip(region)
```

```python
# crs = ee.Projection.crs(3413)
# dir(crs)
# print(crs.name())
# print(crs)
# image
```

```python

Map = geemap.Map()#center=(x, y), zoom=12)

visParams = {'bands': ['B4', 'B3', 'B2'], 'max': .3}
Map.addLayer(image, visParams, 'true-color composite')

Map.centerObject(region)
Map
```

# NDI

```python
# Compute the Normalized Difference Vegetation Index (NDVI).
nir = image.select('B5')
red = image.select('B4')

numerator =  nir.subtract(red)
denominator = nir.add(red)
ndvi = numerator.divide(denominator)

ndvi = ndvi.rename('NDVI')
```

```python
Map = geemap.Map()#center=(x, y), zoom=12)

visParams = {'bands': ['B4', 'B3', 'B2'], 'max': .3}
Map.addLayer(image, visParams, 'true-color composite')

visParams = {'bands': ['NDVI'], 'min': -1, 'max': 1, 'palette': ['brown', 'white', 'green']}
Map.addLayer(ndvi, visParams, 'NDVI')

Map.centerObject(region)
Map
```

# Export

```python
import time 

# Export the image, specifying scale and region.
task = ee.batch.Export.image.toDrive(**{
    'image': image,
    'description': 'imageToDrive',
    'folder':'scires_project_2C',
    'scale': 30,
    'region': region.getInfo()['coordinates']
})
task.start()

while task.active():
  print('Continuing task processing... (id: {}).'.format(task.id))
  time.sleep(5)

print('Finished')
```

```python
import geopandas as gpd
import json
```

```python
aoi_gpd = gpd.read_file('../data/durham_field_polygon.gpkg')

# aoi = geemap.geopandas_to_ee(aoi_gpd)
# aoi = aoi.first().geometry()

aoi = ee.Geometry.Point(-80.498, 37.254).buffer(500).bounds()

aoi
```

```python
Map = geemap.Map()#center=(x, y), zoom=12)
Map.addLayer(aoi, {}, "aoi")
Map.centerObject(aoi)
Map
```

```python
date_begin = '2014-01-01'
date_end = '2023-12-31'

# Get Landsat 8 image collection
landsat8_collection = ee.ImageCollection("LANDSAT/LC08/C02/T1_TOA")

# Filter to desired region and date bounds
landsat8_collection = landsat8_collection.filterBounds(aoi)
landsat8_collection = landsat8_collection.filterDate(date_begin, date_end)
landsat8_collection = landsat8_collection.filter(ee.Filter.lt('CLOUD_COVER', 60))

# Eliminate cloudy pixels
def cloudiness(image):
    cloud = ee.Algorithms.Landsat.simpleCloudScore(image).select('cloud')
    cloud = cloud.reduceRegion(reducer=ee.Reducer.mean(), geometry=aoi, scale=30)
    return image.set(cloud)
landsat8_collection = landsat8_collection.map(lambda image: cloudiness(image))
landsat8_collection = landsat8_collection.filter(ee.Filter.lt('cloud', 30))


# Define the NDI function to process each object
def ndi(image):
    ndi = image.normalizedDifference(['B5', 'B4']).rename('NDVI')
    ndi = ndi.reduceRegion(reducer=ee.Reducer.mean(), geometry=aoi, scale=30)
    return image.set(ndi)

# Map the NDI function across the entire featureCollection
landsat8_collection = landsat8_collection.map(lambda image: ndi(image))


# # Select bands: red, green, blue, NIR, SWIR
# landsat8_collection = landsat8_collection.select(['B4', 'B3', 'B2', 'B5', 'B6'])
```

```python
import pandas as pd

dates_list = landsat8_collection.aggregate_array("DATE_ACQUIRED").getInfo()
print(len(dates_list))
cloudiness_list = landsat8_collection.aggregate_array("cloud").getInfo()
print(len(cloudiness_list))
ndi_list = landsat8_collection.aggregate_array("NDVI").getInfo()
print(len(ndi_list))

# dataframe = pd.DataFrame({'date': dates_list, 'cloud': cloudiness_list})
dataframe = pd.DataFrame({'date': dates_list, 'ndi': ndi_list})
dataframe['date'] = pd.to_datetime(dataframe['date'])
```

```python
import matplotlib.pyplot as plt
%matplotlib widget

fig, ax = plt.subplots(layout='constrained')
dataframe.plot.scatter('date', 'ndi', ax=ax)
dataframe.plot('date', 'ndi', ax=ax, lw=.2)
ax.grid()
```
